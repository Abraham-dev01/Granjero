```python
# Usamos el módulo 'math' 

import math

# Datos para las funciones trigonométricas

def datosTr():
  grados = int(input("\nIntroduce un ángulo en grados: "))
  return grados

# Datos para las funciones exponencial y logarítmica

def datos():
  num = int(input("\nIntroduce el valor del exponencial: "))
  return num

# Cálculo del seno

def seno(grados):
  for i in range(1, grados + 1):
    radianes = math.radians(i)
    resultado = math.sin(radianes)
    print(f"El seno de {i} grados es {resultado}")

# Cálculo del coseno

def coseno(grados):
  for i in range(1, grados + 1):
    radianes = math.radians(i)
    resultado = math.cos(radianes)
    print(f"El coseno de {i} grados es {resultado}")

# Cálculo de la tangente

def tangente(grados):
  for i in range(1, grados + 1):
    radianes = math.radians(i)
    resultado = math.tan(radianes)
    print(f"La tangente de {i} grados es {resultado}")

# Cálculo de la función exponencial

def exponencial(num):
  for i in range(1, num + 1):
    resultado = math.exp(i)
    print(f"La función exponencial de {i} es {resultado}")

# Cálculo del logaritmo neperianao

def neperiano(num):
  for i in range(1, num + 1):
    resultado = math.log(i)
    print(f"El logaritmo neperiano de {i} es {resultado}")

# Pedimos al usuario que elija la función que quiere calcular

opcion = int(input("Introduzca el número de la función que desea calcular: "))

if opcion == 1:
  print("\nSeno")
  grados = datosTr()
  seno(grados)

elif opcion == 2:
  print("\nCoseno")
  grados = datosTr()
  coseno(grados)

elif opcion == 3:
  print("\nTangente")
  grados = datosTr()
  tangente(grados)

elif opcion == 4:
  print("\nExponencial")
  num = datos()
  exponencial(num)

elif opcion == 5:
  print("\nLogaritmo neperiano")
  num = datos()
  neperiano(num)
else:
  print("\nNo hagas mieeeegdaaaas")
```
