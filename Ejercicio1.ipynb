```python
# Funci贸n del descuento

def aplicarDescuento(precio, descuento):  
    return precio - precio * descuento/100

# Funci贸n del IVA

def aplicarIva(precio, porcentaje):
  return precio + (precio * porcentaje / 100)

# Funci贸n que recibe una funci贸n y un diccionario

def diccionario_precios(diccionario, funcion):
  total = 0
  for precio, descuento in diccionario.items():
    total += funcion(precio, descuento)
  return total

subtotal = (diccionario_precios({1000:35, 800:25, 70: 10}, aplicarDescuento))
print("Total con descuento: ", subtotal)
print("Total con IVA: ", aplicarIva(subtotal, 16))  
```
